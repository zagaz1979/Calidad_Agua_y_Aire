```python
# Función reutilizable para normalizar municipios
def normalizar_municipios(df, columna='municipio', correcciones=None):
    df[columna] = df[columna].str.strip().str.upper()
    if correcciones:
        df[columna] = df[columna].replace(correcciones)
    return df
```
